```python
import pandas as pd
import numpy as np
import re
import datetime as DT
```

```python
link=r'C:\Users\User\Desktop\Работа\Test Analys\Internal__Data_Analytics_task__1_.csv'
headlines=[]
pre_table=[]

with open(link,'r') as link_table:
    reader=link_table.readlines()
    
    for i in range(len(re.split(";|\\n",reader[0]))-1):
        headlines.append(re.split(";|\\n",reader[0])[i])

    for i in range(len(reader)-1):
        pre_table.append([re.split(";|\\n",reader[i+1])[j] for j in range(len(re.split(";|\\n",reader[i+1]))-1)])

```

```python
for i in range(len(pre_table)):
    for j in range(len(pre_table[i])):
        if type(pre_table[i][j]) is str:
            if len(pre_table[i][j])<5 and pre_table[i][j]!="":
                pre_table[i][j]=int(pre_table[i][j])
            elif pre_table[i][j]=="":
                pre_table[i][j]=0
           
```

```python
for i in range(len(pre_table)):
    data=re.split("-| ",pre_table[i][0])
#     pre_table[i][0]=[DT.datetime(int(data[0]),int(data[1]),int(data[2]),int(data[3]),0,0,0)]
    pre_table[i][0]=[DT.date(int(data[0]),int(data[1]),int(data[2])),DT.time(int(data[3]),0)]
```

```python
week=4
old_date=pre_table[0][0][0]
for i in range(len(pre_table)):
    if week>7:
        week=1
    if old_date==pre_table[i][0][0]:
        pre_table[i].append(week)
    elif old_date!=pre_table[i][0][0]:
        week+=1
        if week>7:
            week=1
        pre_table[i].append(week)
        old_date=pre_table[i][0][0]
        
headlines.append("Weekday")
```
